```python
import pandas as pd
import numpy as np
```

```python
AT1 = pd.read_excel('AT.Datos globales de la Oferta en apartamentos turísticos en la ciudad de Madrid por Mes.xlsx')
AT2 = pd.read_excel('AT.Pernoctaciones en la ciudad de Madrid por País de Residencia y Mes.xlsx')
AT3 = pd.read_excel('AT.Viajeros en la ciudad de Madrid por País de Residencia y Mes.xlsx')
AT4 = pd.read_excel('AT.Viajeros y pernoctaciones en la ciudad de Madrid por Mes.xlsx')
  
```

```python
AT1.columns=[e.lower().replace(' ', '_') for e in AT1.columns] 
display(AT1.head())
AT2.columns=[e.lower().replace(' ', '_') for e in AT2.columns] 
display(AT2.head())
AT3.columns=[e.lower().replace(' ', '_') for e in AT3.columns] 
display(AT3.head())
AT4.columns=[e.lower().replace(' ', '_') for e in AT4.columns] 
display(AT4.head())
```

```python
AT1['date'] = AT1['mes'].astype(str) +"-"+ AT1['año'].astype(str)
AT2['date'] = AT2['mes'].astype(str) +"-"+ AT2['año'].astype(str)
AT3['date'] = AT3['mes'].astype(str) +"-"+ AT3['año'].astype(str)
AT4['date'] = AT4['mes'].astype(str) +"-"+ AT4['año'].astype(str)
```

```python
AT1.columns
```

```python
AT_totals= AT1[['date', '#_turistic_apartments', 'ta_capacity',
       'ta_occupancy_per_capacity', 'ta_occupancy_per_apartment',
       'ta_occupancy_per_capacity_during_weekends', 'ta_staff']]
```

```python
AT2.columns
```

```python
AT2_2=AT2[['date','año', 'mes', 'ta_nights_stays_alemania', 'ta_nights_stays_austria',
       'ta_nights_stays_bélgica', 'ta_nights_stays_dinamarca',
       'ta_nights_stays_finlandia', 'ta_nights_stays_francia',
       'ta_nights_stays_grecia', 'ta_nights_stays_irlanda',
       'ta_nights_stays_italia', 'ta_nights_stays_luxemburgo',
       'ta_nights_stays_países_bajos', 'ta_nights_stays_polonia',
       'ta_nights_stays_portugal', 'ta_nights_stays_república_checa',
       'ta_nights_stays_suecia', 'ta_nights_stays_resto_de_unión_europea',
       'ta_nights_stays_noruega', 'ta_nights_stays_reino_unido',
       'ta_nights_stays_rusia', 'ta_nights_stays_suiza',
       'ta_nights_stays_resto_de_europa', 'ta_nights_stays_estados_unidos',
       'total_ta_nights_stays_central_and_south_america',
       'ta_nights_stays_africa', 'ta_nights_stays_resto_del_mundo']]

AT3_2=AT3.drop(['mes','año','date'],axis=1)
AT_countries=pd.concat([AT2_2, AT3_2], axis = 1)

```

```python
AT4.columns
```

```python
AT4.columns
AT4_2=AT4[['ta_total_travelers', 'ta_total_spanish_travelers',
       'ta_total_foreign_travelers', 'ta_total_nigts_stays',
       'ta_total_nights_stays_spanishs', 'ta_total_nights_stays_foreigns',
       'ta_average_stay_(days)']]
AT_countries=pd.concat([AT_countries, AT4_2], axis = 1)
```

```python
AT_totals.shape
```

```python
AT_countries.shape
```

```python
AT_df= pd.merge(left = AT_totals,
                                 right = AT_countries,
                                 how = 'left', 
                                 left_on = "date", 
                                 right_on= "date")
AT_df
```

```python
AT_df.columns
```

```python
AT_df= AT_df[['date', 'año', 'mes', '#_turistic_apartments', 'ta_capacity',
       'ta_occupancy_per_capacity', 'ta_occupancy_per_apartment',
       'ta_occupancy_per_capacity_during_weekends', 'ta_staff', 
       'ta_nights_stays_alemania', 'ta_nights_stays_austria',
       'ta_nights_stays_bélgica', 'ta_nights_stays_dinamarca',
       'ta_nights_stays_finlandia', 'ta_nights_stays_francia',
       'ta_nights_stays_grecia', 'ta_nights_stays_irlanda',
       'ta_nights_stays_italia', 'ta_nights_stays_luxemburgo',
       'ta_nights_stays_países_bajos', 'ta_nights_stays_polonia',
       'ta_nights_stays_portugal', 'ta_nights_stays_república_checa',
       'ta_nights_stays_suecia', 'ta_nights_stays_resto_de_unión_europea',
       'ta_nights_stays_noruega', 'ta_nights_stays_reino_unido',
       'ta_nights_stays_rusia', 'ta_nights_stays_suiza',
       'ta_nights_stays_resto_de_europa', 'ta_nights_stays_estados_unidos',
       'total_ta_nights_stays_central_and_south_america',
       'ta_nights_stays_africa', 'ta_nights_stays_resto_del_mundo',
       'ta_travelers_alemania', 'ta_travelers_austria', 'ta_travelers_bélgica',
       'ta_travelers_dinamarca', 'ta_travelers_finlandia',
       'ta_travelers_francia', 'ta_travelers_grecia', 'ta_travelers_irlanda',
       'ta_travelers_italia', 'ta_travelers_luxemburgo',
       'ta_travelers_países_bajos', 'ta_travelers_polonia',
       'ta_travelers_portugal', 'ta_travelers_república_checa',
       'ta_travelers_suecia', 'ta_travelers_resto_de_unión_europea',
       'ta_travelers_noruega', 'ta_travelers_reino_unido',
       'ta_travelers_rusia', 'ta_travelers_suiza',
       'ta_travelers_resto_de_europa', 'ta_travelers_estados_unidos',
       'ta_travelers_resto_de_américa', 'ta_travelers_africa',
       'ta_travelers_resto_del_mundo', 'ta_total_travelers',
       'ta_total_spanish_travelers', 'ta_total_foreign_travelers',
       'ta_total_nigts_stays', 'ta_total_nights_stays_spanishs',
       'ta_total_nights_stays_foreigns', 'ta_average_stay_(days)']]
```

```python
from datetime import datetime
AT_df['date'] = pd.to_datetime(AT_df['date'])
AT_df['date'] = AT_df['date'].apply(lambda x: datetime.strftime(x,'%Y-%m-%d'))

AT_df['date'].head()
```

```python
AT_df.to_csv('TA_df.csv', index=False)
```
