```python
import pandas as pd
import numpy as np
```

```python
AT1 = pd.read_excel('AT.Datos globales de la Oferta en apartamentos turísticos en la ciudad de Madrid por Mes.xlsx')
AT2 = pd.read_excel('AT.Pernoctaciones en la ciudad de Madrid por País de Residencia y Mes.xlsx')
AT3 = pd.read_excel('AT.Viajeros en la ciudad de Madrid por País de Residencia y Mes.xlsx')
```

```python
AT1.columns=[e.lower().replace(' ', '_') for e in AT1.columns] 
display(AT1.head())
AT2.columns=[e.lower().replace(' ', '_') for e in AT2.columns] 
display(AT2.head())
AT3.columns=[e.lower().replace(' ', '_') for e in AT3.columns] 
display(AT3.head())
```

```python
AT1['date'] = AT1['mes'].astype(str) +"-"+ AT1['año'].astype(str)
AT2['date'] = AT2['mes'].astype(str) +"-"+ AT2['año'].astype(str)
AT3['date'] = AT3['mes'].astype(str) +"-"+ AT3['año'].astype(str)
```

```python
AT1.columns
AT_totals= AT1[['date', 'mes','año', 'número_plazas_', 'número_apartamentos_estimados',
       'grado_ocupación_por_plazas_(%)',
       'grado_ocupación_por_apartamentos_(%)',
       'grado_ocupación_por_apartamentos_fin_de_semana_(%)',
       'personal_ocupado_(número)']]
```

```python
AT2_2=AT2[['date','año', 'mes', 'total', 'pernoctaciones_nacionales',
       'pernoctaciones_extranjeros', 'pernoctaciones_alemania',
       'pernoctaciones_austria', 'pernoctaciones_bélgica',
       'pernoctaciones_dinamarca', 'pernoctaciones_finlandia',
       'pernoctaciones_francia', 'pernoctaciones_grecia',
       'pernoctaciones_irlanda', 'pernoctaciones_italia',
       'pernoctaciones_luxemburgo', 'pernoctaciones_países_bajos',
       'pernoctaciones_polonia', 'pernoctaciones_portugal',
       'pernoctaciones_república_checa', 'pernoctaciones_suecia',
       'pernoctaciones_resto_de_unión_europea', 'pernoctaciones_noruega',
       'pernoctaciones_reino_unido', 'pernoctaciones_rusia',
       'pernoctaciones_suiza', 'pernoctaciones_resto_de_europa',
       'pernoctaciones_estados_unidos', 'pernoctaciones_resto_de_américa',
       'pernoctaciones_países_africanos', 'pernoctaciones_resto_del_mundo']]

AT3_2=AT3.drop(['mes','año','date'],axis=1)
AT_countries=pd.concat([AT2_2, AT3_2], axis = 1)

```

```python
AT_totals.shape
```

```python
AT_countries.shape
```

```python
AT_df= pd.merge(left = AT_totals,
                                 right = AT_countries,
                                 how = 'left', 
                                 left_on = "date", 
                                 right_on= "date")
AT_df
```

```python
AT_df=AT_df.drop(['mes_y','año_y'],axis=1)
AT_df_c=list(AT_df.columns)
AT_df_c
```

```python
AT_df= AT_df.rename(columns={'mes_x':'mes', 'año_x':'año'})
AT_df_c=list(AT_df.columns)
AT_df_c
```
