```python
import pandas as pd
import numpy as np
```

```python
EV = pd.read_excel('IFEMA.xlsx')
```

```python
EV.columns=[e.lower().replace(' ', '_') for e in EV.columns] 
display(EV.head())
```

```python
EV['total_eventos']=EV['exposición']+EV['congresos_y_convenciones'] +EV['conciertos_y_festivales']
```

```python
EV=EV[['date','mes','año', 'total_eventos', 'exposición',
       'congresos_y_convenciones', 'conciertos_y_festivales',
       'total_expositores', 'expositores_nacionales',
       'expositores_extranjeros', 'total_visitantes',
       'total_visitantes_profesionales', 'visitantes_profesionales_nacionales',
       'visitantes_profesionales_extranjero', 'visitantes_particulares',
       'número_de_días_feriales']]
EV
```
